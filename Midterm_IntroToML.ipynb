#### Check Duplicated Rows

```python
# check duplicated rows

duplicated_rows_cnt = df.duplicated().sum()
print(f"Number of duplicated rows: {duplicated_rows_cnt}")
```

```python
# Outlier (price) removal

def remove_outliers_percentile(df, column, lower_percentile=0.01, upper_percentile=0.99):
    lower_bound = df[column].quantile(lower_percentile)
    upper_bound = df[column].quantile(upper_percentile)
    filtered_df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

    return filtered_df

# Function to remove outliers using Z-score
def remove_outliers_zscore(df, column, z_thresh=3):
    z_scores = np.abs(stats.zscore(df[column]))
    filtered_df = df[z_scores < z_thresh]

    return filtered_df

```

#### Remove Outliers

```python
# Removing outliers

print(f"Number of rows before outlier removal: {df.shape[0]}")
df = remove_outliers_percentile(df, 'price', lower_percentile=0.01, upper_percentile=0.99)
print(f"Number of rows after outlier removal: {df.shape[0]}")
df = remove_outliers_zscore(df, 'price', z_thresh=3)
print(f"Number of rows after outlier removal: {df.shape[0]}")
```
